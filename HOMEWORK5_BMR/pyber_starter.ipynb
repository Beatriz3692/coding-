```python
%matplotlib inline
# Dependencies and Setup
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np

# File to Load (Remember to change these)
city_data = "data/city_data.csv"
ride_data = "data/ride_data.csv"

# Display the data table for preview
```

```python
# Read the City and Ride Data
city_df = pd.read_csv(city_data)
ride_df = pd.read_csv(ride_data)

# Combine the data into a single dataset
merge_table = pd.merge(city_df, ride_df)

# Display the data table for preview
merge_table.head()
```

## Bubble Plot of Ride Sharing Data

```python
# Obtain the x and y coordinates for each of the three city types
urban_df = merge_table.loc[merge_table['type'] == "Urban"]
suburban_df = merge_table.loc[merge_table['type'] == "Suburban"]
rural_df = merge_table.loc[merge_table['type'] == "Rural"]

urban_city = urban_df.groupby("city")
suburban_city = suburban_df.groupby("city")
rural_city = rural_df.groupby("city")
```

```python
# x
urban_x = urban_city["ride_id"].count()
suburban_x = suburban_city["ride_id"].count()
rural_x = rural_city["ride_id"].count()

# y 
urban_rides = urban_city["city"].count()
suburban_rides = suburban_city["city"].count()
rural_rides = rural_city["city"].count()

urban_fares = urban_city["fare"].sum()
suburban_fares = suburban_city["fare"].sum()
rural_fares = rural_city["fare"].sum()

urban_avg_fare = urban_fares/urban_rides
suburban_avg_fare = suburban_fares/suburban_rides
rural_avg_fare = rural_fares/rural_rides
```

```python
# Build the scatter plots for each city types
urban_plt = plt.scatter(urban_x, urban_avg_fare, marker="o", facecolors="coral", edgecolors="black", s=urban_fares, alpha=0.75, label="Urban")
suburban_plt = plt.scatter(suburban_x, suburban_avg_fare, marker="o", facecolors="skyblue", edgecolors="black", s=suburban_fares, alpha=0.75, label="Suburban")
rural_plt = plt.scatter(rural_x, rural_avg_fare, marker="o", facecolors="yellow", edgecolors="black", s=rural_fares, alpha=0.75, label="Rural")

plt.legend(handles=[urban_plt, suburban_plt, rural_plt], loc="best")
plt.grid()
plt.title("Pyber Ride Sharing Data (2016)")
plt.xlabel("Total Number of Rides (Per City)")
plt.ylabel("Average Fare ($)")
plt.show()
```

## Total Fares by City Type

```python
# Build Pie Chart
labels = ["Rural", "Suburban", "Urban"]
sizes = merge_table.groupby('type')['fare'].sum()
colors = ["yellow", "lightblue", "coral"]
explode = (0, 0, 0.1)
```

```python

plt.pie(sizes, explode=explode, labels=labels, colors=colors,  autopct="%1.1f%%", shadow=True, startangle=140)
plt.title('% of Total Fares by City Type')
plt.axis("equal")
plt.show()
```

## Total Rides by City Type

```python
# Calculate Ride Percents
ride_sizes = merge_table.groupby('type')['ride_id'].count()

# Build Pie Chart
plt.pie(ride_sizes, explode=explode, labels=labels, colors=colors,  autopct="%1.1f%%", shadow=True, startangle=140)
plt.title('% of Total Rides by City Type')
plt.axis("equal")
plt.show()
# Save Figure
plt.savefig("../Images/TotalRides.png")
```

## Total Drivers by City Type

```python
# Calculate Driver Percents
drivers = merge_table.groupby(['type','city'])['driver_count'].first()
drivers.reset_index()
drivers_size = drivers.groupby('type').sum()

# Build Pie Charts
plt.pie(drivers_size, explode=explode, labels=labels, colors=colors,  autopct="%1.1f%%", shadow=True, startangle=140)
plt.title('% of Total Drivers by City Type')
plt.axis("equal")
plt.show()
# Save Figure
plt.savefig("../Images/TotalDrivers.png")
```
