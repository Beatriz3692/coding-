# Reflect Tables into SQLAlchemy ORM

```python
# Python SQL toolkit and Object Relational Mapper
import sqlalchemy
from sqlalchemy.ext.automap import automap_base
from sqlalchemy.orm import Session
from sqlalchemy import create_engine, func
```

```python
engine = create_engine("sqlite:///Resources/hawaii.sqlite")
```

```python
# reflect an existing database into a new model
Base = automap_base()
# reflect the tables
Base.prepare(engine, reflect=True)
```

```python
# We can view all of the classes that automap found
Base.classes.keys()
```

```python
# Save references to each table
Measurement = Base.classes.measurement
Station = Base.classes.station
```

```python
# Create our session (link) from Python to the DB
session = Session(engine)
```

# Exploratory Climate Analysis

```python
# Design a query to retrieve the last 12 months of precipitation data and plot the results

data = session.query(Measurement.date).all()
for result in data:
    print(result)
    
# Use Pandas Plotting with Matplotlib to plot the data
```

```python
#access all precipitation data from the last year 
start_date = dt.date(2019, 8, 5)
end_date = dt.date(2019, 8, 17)

prcp_data = session.query(func.strftime("%Y-%m-%d", Measurement.date), Measurement.prcp).\
    filter(func.strftime("%Y-%m-%d", Measurement.date) >= dt.date(2017, 8, 17)).all()
for results in prcp_data:
    print(results)
```

```python
# Save the query results as a Pandas DataFrame and set the index to the date column
prcp_df = pd.DataFrame(prcp_data, columns= ["date", "precipitation"])
# Sort the dataframe by date
prcp_df.set_index('date', inplace=True)
prcp_df.head(10)
```

```python
# Use Pandas Plotting with Matplotlib to plot the data
fig, ax = plt.subplots(figsize = (15, 7))
prcp_df.plot(ax = ax, x_compat = True)

# set x and y labels
ax.set_xlabel('Date')
ax.set_ylabel('Precipitation')
plt.title( "Hawaii Precipitation '17")
plt.tight_layout()
plt.show()

plt.savefig("Images/prcp.png")
```

```python
# Use Pandas to calcualte the summary statistics for the precipitation data
prcp_df.describe()
```

# Station Analysis

```python
# Design a query to show how many stations are available in this dataset?
stations = session.query(Station).count()
print(stations)
```

```python
# What are the most active stations? (i.e. what stations have the most rows)?
# List the stations and the counts in descending order.
stations = session.query(Station.station, func.count(Measurement.id)).select_from(Measurement).\
    join(Station, Measurement.station == Station.station).group_by(Station.station).\
    order_by(func.count(Measurement.id).desc()).all()
print(stations)
```

```python
# most active 
station_ID = stations[0][0]
most_active = stations[0][0:2]
```

```python
most_active
```

```python
# Using the station id from the previous query, calculate the
#lowest temperature recorded, 
lowest_temp = (session.query(Measurement.tobs)
                      .filter(Measurement.station == station_ID).order_by(Measurement.tobs.asc())
                     .first())
print (f"The lowest temperature: {lowest_temp} Farenheit")
# highest temperature recorded
highest_temp = (session.query(Measurement.tobs)
                      .filter(Measurement.station == station_ID).order_by(Measurement.tobs.desc())
                     .first())
print (f"The highest temperature: {highest_temp} Farenheit")
#and average temperature most active station?
avg_temp = (session.query(func.avg(Measurement.tobs))
                  .filter(Measurement.station == station_ID)).all()
avg_temp ='{0:.3}'.format(avg_temp[0][0])
print (f"The average temperature: {avg_temp} Farenheit")
```

```python
# Choose the station with the highest number of temperature observations.
# Query the last 12 months of temperature observation data for this station and plot the results as a histogram
highest_temp_obs = session.query(Measurement.station, Measurement.date, Measurement.tobs).\
    filter(Measurement.date > '2017-01-01').\
    order_by(Measurement.date).all()
highest_temp_obs
```

```python
# create dataframe
temp_df = pd.DataFrame(highest_temp_obs, columns=['Station', 'date', 'temperature'])
temp_df.set_index('Station', inplace=True)
temp_df.head()
```

```python
# create histogram
temp_plot = temp_df['temperature'].hist(bins=12, figsize=(15,10))
temp_plot.set_ylabel('Frequency')
temp_plot.set_xlabel('Temperature')
temp_plot.set_title('Temperature Observations of 2017')
plt.savefig("Images/temp.png")
```

# Climate App 

```python
# use calc temp to find tmin, tavg, tmax
def calc_temps(start_date, end_date):
    "return the minimum, average, and maximum temperatures for that range of date"
    
    """Args:
        start_date (string): A date string in the format %Y-%m-%d
        end_date (string): A date string in the format %Y-%m-%d
        """"""
    """
    
    return session.query(func.min(Measurement.tobs), func.avg(Measurement.tobs), func.max(Measurement.tobs)).\
        filter(Measurement.date >= start_date).filter(Measurement.date <= end_date).all()

# function using trip dates
t_trip = calc_temps('2017-08-05', '2017-08-17')
t_trip
```

```python
t_min=t_trip [0][0]
t_avg=t_trip [0][1]
t_max=t_trip [0][2]
```

```python
# Plot the results from your previous query as a bar chart.
fig = plt.figure()
# Use the peak-to-peak (tmax-tmin) value as the y error bar (yerr)
plt.bar( 1,t_avg, color = 'pink', yerr = t_max-t_min, align='center')
plt.xlim(-0.001, 2)
plt.ylim([0, 100])
# Use the average temperature for the y value
plt.ylabel("Temperature (F)")
# Use "Trip Avg Temp" as your Title
plt.title('"Trip Avg Temp"')
fig.savefig("Images/avg_temp.png")
```
